```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data from csv file
data = pd.read_csv('csv.csv')
data.head()
```

```python
len(data)
```

```python
# plot Value
plt.plot(data['Value'])
# x label is Episode
plt.xlabel('Episode')
# y label is Mean Reward
plt.ylabel('Mean Reward')
# title is Mean Reward vs Episode
plt.title('Mean Reward vs Episode')
# draw horizontal line at y = 0
# plt.axhline(y=0, color='r', linestyle='-', label='Zero')
# draw vertical line where we have max value
# calculate max value and where is it
max_value = data['Value'].max()
max_index = data['Value'].idxmax()
# draw vertical line
plt.axvline(x=max_index, color='orange', linestyle='--', label='Max Value')
# draw moving average
plt.plot(data['Value'].rolling(20).mean(), label='Moving Average', color='r')
# add grid
plt.grid()
# show legend
plt.legend()
# save figure
plt.savefig('plot.png')
```
