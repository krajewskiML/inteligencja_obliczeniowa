```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# PPO
## Discount factor = 0.99

```python
ppo_99 = pd.read_csv('ppo_99.csv')
ppo_99.head()
```

```python
# plot values but on the x axis we have the number of episodes multiplied by 2
plt.plot(ppo_99['Step'] / 1024, ppo_99['Value'])
```

## Discount factor = 0.9

```python
ppo_9 = pd.read_csv('ppo_9.csv')
ppo_9.head()
```

```python
# plot values but on the x axis we have the number of episodes multiplied by 2
plt.plot(ppo_9['Step'] / 1024, ppo_9['Value'])
```

### Discount factor = 0.95

```python
ppo_95 = pd.read_csv('ppo_95.csv')
ppo_95.head()
```

```python
# plot values but on the x axis we have the number of episodes multiplied by 2
plt.plot(ppo_95['Step'] / 1024, ppo_95['Value'])
```

### Compare all

```python
# plot values but on the x axis we have the number of episodes multiplied by 2
# we want to show all the values on the same plot
plt.plot(ppo_99['Step'] / 1024, ppo_99['Value'], label='gamma = 0.99')
plt.plot(ppo_9['Step'] / 1024, ppo_9['Value'], label='gamma = 0.9')
plt.plot(ppo_95['Step'] / 1024, ppo_95['Value'], label='gamma = 0.95')
plt.legend()
# save figure

plt.xlabel('Steps [x1000]')
plt.ylabel('Rewards')
plt.savefig('ppo.png')
```

# DQN
## Discount factor = 0.99

```python
dqn_99 = pd.read_csv('dqn_99.csv')
dqn_99.head()
```

```python
# plot values but on the x axis we have the number of episodes multiplied by 2
plt.plot(dqn_99['Step'] / 201 / 5, dqn_99['Value'])
# add vertical lines at episode 50 and name it exploration
plt.axvline(x=50, color='r', linestyle='--')
plt.text(50, 0.5, 'exploration', rotation=270)
# save figure
plt.savefig('dqn_99.png')
```

## Discount factor = 0.95

```python
dqn_95 = pd.read_csv('dqn_95.csv')
dqn_95.head()
```

# Compare PPO and DQN

```python
# compare the two algorithms
# we want to show all the values on the same plot
plt.plot(dqn_99['Step'] / 201 / 5, dqn_99['Value'], label='DQN')
plt.plot(ppo_99['Step'] / 1024, ppo_99['Value'], label='PPO')
plt.legend()
# add vertical lines at episode 50 and name it exploration
plt.axvline(x=50, color='r', linestyle='--')
plt.text(50, 0.5, 'exploration', rotation=270)

plt.xlabel('Steps [x1000]')
plt.ylabel('Rewards')
# save figure
plt.savefig('dqn_vs_ppo.png')
```
